```python
import pandas as pd
data = pd.read_csv(filePath)
```

```python
data['lineage'] = np.where(data['lineage_2'].str.contains('Non-small', case=False), 'Non-Small Cell', 'Small Cell')
```

```python
data['lineage'].value_counts()
```

```python
columns_to_remove = ['depmap_id', 'cell_line_display_name', 'lineage_1', 'lineage_2', 'lineage_3', 'lineage_5', 'lineage_6', 'lineage_4']
data = data.drop(columns=columns_to_remove)
```

```python
data.info()
```

```python
any_na_in_rows = data.isna().any(axis=1).sum() > 0

if any_na_in_rows:
    print("There are NA values in some rows.")
else:
    print("There are no NA values in any row.")
```

```python
any_na_in_columns = data.isna().any()

if any_na_in_columns.any():
    print("There are NA values in some columns.")
    print("Columns with NA values:", data.columns[data.isna().any()].tolist())
else:
    print("There are no NA values in any column.")
```

```python
data.head(2)
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
y_train.value_counts()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(n_estimators=100, random_state=42)
forest_clf.fit(X_train_scaled, y_train)
```

```python
y_pred = forest_clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate precision, recall, and F1 score
precision1 = precision_score(y_test, y_pred, pos_label='Non-Small Cell')
precision2 = precision_score(y_test, y_pred, pos_label='Small Cell')

recall1 = recall_score(y_test, y_pred, pos_label='Non-Small Cell')
recall2 = recall_score(y_test, y_pred, pos_label='Small Cell')

f1_1 = f1_score(y_test, y_pred, pos_label='Non-Small Cell')
f1_2 = f1_score(y_test, y_pred, pos_label='Small Cell')

print("Precision for Non-Small Cell:", precision1)
print("Precision for Small Cell:", precision2)

print("Recall for Non-Small Cell:", recall1)
print("Recall for Small Cell:", recall2)

print("F1 Score for Non-Small Cell:", f1_1)
print("F1 Score for Small Cell:", f1_2)
```

```python
importances = forest_clf.feature_importances_

# Sort feature importances in descending order
sorted_indices = np.argsort(importances)[::-1]

# Cumulative sum of feature importances
cumulative_importances = np.cumsum(importances[sorted_indices])

# Find the number of features that account for 90% of prediction
num_features_90_percent = np.argmax(cumulative_importances >= 0.9) + 1

print("Number of features accounting for 90% of prediction:", num_features_90_percent)
```
